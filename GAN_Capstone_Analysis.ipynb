## Plots by epoch for Gaussian Pulse

```python
# function to prepare data to be plotted
def safety_glitch_plot_prep(x, epoch_index, batch_index=None):
  if batch_index == None:
    batch_index = random.randint(0, 31)
  return np.squeeze(x[epoch_index][batch_index])
```

```python
# random seed for control
random.seed(419)


fig, axs = plt.subplots(3, 5, figsize=(30,15))
# epoch 1
axs[0, 0].set_title('Epoch 1')
axs[0,0].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=1), 'tab:orange')
axs[1,0].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=1), 'tab:orange')
axs[2,0].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=1), 'tab:orange')
# epoch 5
axs[0, 1].set_title('Epoch 5')
axs[0,1].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=5), 'tab:blue')
axs[1,1].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=5), 'tab:blue')
axs[2,1].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=5), 'tab:blue')
# epoch 200
axs[0, 2].set_title('Epoch 200')
axs[0,2].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=200), 'tab:green')
axs[1,2].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=200), 'tab:green')
axs[2,2].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=200), 'tab:green')
# epoch 300
axs[0, 3].set_title('Epoch 300')
axs[0,3].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=300), 'tab:red')
axs[1,3].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=300), 'tab:red')
axs[2,3].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=300), 'tab:red')
# epoch 400
axs[0, 4].set_title('Epoch 500')
axs[0,4].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=499), 'tab:purple')
axs[1,4].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=499), 'tab:purple')
axs[2,4].plot(safety_glitch_plot_prep(gaussian_safety_glitches, epoch_index=499), 'tab:purple')

# title
fig.suptitle('Gaussian Pulse Generated Samples Across Epochs', fontsize=40)
fig.text(0.5, 0.04, 'Time', ha='center', fontsize=25)
fig.text(0.04, 0.5, 'Frequency', va='center', rotation='vertical', fontsize=25)

plt.savefig('/content/gdrive/MyDrive/GAN/gaussian_pulse_generated_across_epochs.png')
```

## Loss plots for gaussian, sine, and gengli's gaussian

```python
# function to prep data used for plotting loss
def loss_plot_prep(data, disc_error_list, gen_error_list, disc_error_gp_list,
                   disc_error_ct_list):
  for key, value in data.items():
    disc_error_list.append(value['discriminator_error'])
    gen_error_list.append(value['generator_error'])
    disc_error_gp_list.append(value['discriminator_error_gp'])
    disc_error_ct_list.append(value['discriminator_error_ct'])

# fixed list of epochs since all GANs have the same number of epochs
epochs = list(range(500))
```

```python
# data prep for gaussian GAN
gaussian_disc_error = list()
gaussian_gen_error = list()
gaussian_disc_error_gp = list()
gaussian_disc_error_ct = list()

loss_plot_prep(data=gaussian_data, disc_error_list=gaussian_disc_error,
               gen_error_list=gaussian_gen_error,
               disc_error_gp_list=gaussian_disc_error_gp,
               disc_error_ct_list=gaussian_disc_error_ct)
```

```python
# plots for gaussian GAN loss

fig, axs = plt.subplots(3, 1, figsize=(20,12))
# discriminator and generator
axs[0].set_title('Discriminator and Generator Error')
axs[0].plot(epochs, gaussian_disc_error, 'tab:orange', label='Discriminator Error')
axs[0].plot(epochs, gaussian_gen_error, 'tab:blue', label='Generator Error')
axs[0].legend()
# gradient penalty
axs[1].set_title('Discriminator Gradient Penalty (GP)')
axs[1].plot(epochs, gaussian_disc_error_gp, 'tab:green', label='GP')
axs[1].legend()
# consistency term
axs[2].set_title('Discriminator Consistency Term (CT)')
axs[2].plot(epochs, gaussian_disc_error_ct, 'tab:purple', label='CT')
axs[2].legend()


# title
fig.suptitle('Gaussian Pulse GAN Loss Across Epochs', fontsize=40)
fig.text(0.5, 0.04, 'Epoch', ha='center', fontsize=25)
fig.text(0.04, 0.5, 'Loss', va='center', rotation='vertical', fontsize=25)

plt.savefig('/content/gdrive/MyDrive/GAN/gaussian_loss_across_epochs.png')
```

```python
# data prep for sine GAN
sine_disc_error = list()
sine_gen_error = list()
sine_disc_error_gp = list()
sine_disc_error_ct = list()

loss_plot_prep(data=sine_data, disc_error_list=sine_disc_error,
               gen_error_list=sine_gen_error,
               disc_error_gp_list=sine_disc_error_gp,
               disc_error_ct_list=sine_disc_error_ct)
```
